```python
from filtros import *
from functions import *
from fit_voc import *
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
def log_a(a,x):
    y = np.log(x)/np.log(a)
    return y
```

```python
SR =         ["100-0    ",
              "100-25   ",
              "75-0     ",
              "100-50   ",
              "75-25    ",
              "50-0     ",
              "100-75   ",
              "75-50    ",
              "62.5-37.5",
              "50-25    ",
              "25-0     ",]

percentage =  [1.00000,
            0.78750,
            1.12525,
            0.43750,
            0.68750,
            1.03125,
            0.40625,
            0.29700,
            0.28125,
            0.62500,
            1.00000]

base_data = list(zip(SR,percentage))
base_headers = ["SR","Percent"]
base_table = pd.DataFrame(base_data,columns=base_headers)

```

### Simulamos el caso de una celda nueva

```python
# Datos del Datasheet
new_EOL = 0.7
new_cycles = 500
new_SR = 0.8

# Obtenemos los ciclos equivalentes de la nueva celda
new_cycles_eq = 500*new_SR
# Calculamos el eta y los ciclos para el EOL de la tabla proporcional
new_eta_0 = new_EOL**(1/new_cycles_eq)
# Obtenemos los ciclos equivalentes para el EOL original de la tabla
new_cycles_eq_08 = 1/log_a(0.8,new_eta_0)
```

```python
EQ_cycles = []
for percentage in base_table.Percent.values:
    eq_cycle_k = percentage*new_cycles_eq_08
    # eq_cycle_k = percentage*500
    EQ_cycles.append(eq_cycle_k)
base_table["EQ_cycles"] = EQ_cycles

base_table
```

```python
# agregamos los subciclos que induce cada factor
cycles_0 = 500
EOL_0 = 0.7
eta_0 = EOL_0**(1/cycles_0)
EQ_cycles = []

for percentage in base_table.Percent.values:
    eq_cycle_k = percentage*cycles_0
    EQ_cycles.append(eq_cycle_k)
base_table["EQ_cycles_validacion"] = EQ_cycles

base_table
```

### Ahora calculamos los valorea de eta

```python
eta_values = []
for cycle in base_table.EQ_cycles_validacion.values:
    eta_k = 0.7**(1/cycle)
    eta_values.append(eta_k)
base_table["Eta_EQ_cycles_validacion"] = eta_values

eta_values = []
for cycle in base_table.EQ_cycles.values:
    eta_k = 0.8**(1/cycle)
    eta_values.append(eta_k)
base_table["Eta_EQ_cycles"] = eta_values

base_table
```
