```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import csv
```

Importamos los datos del gráfico de temperatura v/s vida útil

```python
df = df = pd.read_csv('graph_data.csv', names=['x', 'y'],delimiter=';', skiprows=1)

# Assuming the columns are named 'x' and 'y'
x_values = np.array(df['x'])
y_values = np.array(df['y'])
y_values = y_values/2000
```

```python
plt.plot(x_values,y_values)
plt.show()
```

Hacemos el fiteo del polinomio

```python
# Fit a 6th-degree polynomial to the data
coefficients = np.polyfit(x_values, y_values, 6)

# Create a polynomial function from the coefficients
polynomial = np.poly1d(coefficients)

# Generate x values from -40 to 80
x_range = np.linspace(-40, 80, 120)  # 500 points for a smooth curve

# Calculate the corresponding y values using the polynomial function
y_range = polynomial(x_range)

# Plot the polynomial function
plt.plot(x_range, y_range, label="6th-degree Polynomial")
plt.scatter(x_values, y_values, color='red', label="Data Points")  # Optional: Plot original data points
plt.title("6th-degree Polynomial Fit")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
